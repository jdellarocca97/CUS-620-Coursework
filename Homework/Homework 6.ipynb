```python
import numpy as np
import scipy as sp
import pandas as pd

def get_input_descriptor():
    a = input("What file do you want to open and read?: ")
    try:
        f=open(a.lower()+ ".csv")
        b=pd.read_csv(f)
        x=pd.DataFrame(b)
        print("Here are the first 5 rows:\n")
        print(x.head())
        return(x)
    except:
        print("Not a file name.  Try again.")
        get_input_descriptor()        
        
def get_data_list(x):
    b=int(input("\nWhat column do you want to check?: "))
    if b == 0:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==1:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["Open"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==2:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["High"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==3:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["Low"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==4:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["Close"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==5:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["Volume"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    elif b==6:
        w=list()
        for i in range(len(x)):
            z=list()
            z.append(x["Date"][i])
            z.append(x["Adj Close"][i])
            z=tuple(z)
            w.append(z)
        return(w)
    else:
        print("That is not a column.  Try Again.")
        
def average_data(l):
    a = {}
    counter = {}
    for i in l:
        b=i[0].split("-")
        b=b[1]+"-"+b[0]
        if b in a:
            a[b] += i[1]
            counter[b] += 1
        else:
            a[b] = i[1]
            counter[b] = 1
    avg = []
    for key in a:
        avg.append((round(a[key]/counter[key],2), key))
    sort_avg = sorted(avg)
    print("\nLowest 6 for column 6:")
    for item in sort_avg[0:6]:
        print("Date: {}, Value: {}".format(item[1],item[0]))
    revsort_avg = sorted(avg, reverse=True)    
    print("\nHighest 6 for column 6:")
    for item in revsort_avg[0:6]:
        print("Date: {}, Value: {}".format(item[1],item[0]))
      
        
def close_file():
    try:
        f.close()
        print("\nThis file has closed.")
    except:
        print("\nThere was an error closing this file.")
    
def main():
    try:
        f=get_input_descriptor()
        l=get_data_list(f)  
        average_data(l)
        close_file()
    except:
        print("\nThere was an error.  Try again.")
        main()
main()
```
