```python
#!/usr/bin/env python3

import csv
import sys

# a file in the current directory
FILENAME = "monthly_sales.csv"

def read_sales():
    try:
        sales = []
        with open(FILENAME, newline="") as file:
            reader = csv.reader(file)
            for row in reader:
                sales.append(row)
        return sales
        
    except FileNotFoundError:
        print("Could not find " + FILENAME + " file.")
        print("Exiting program. Bye!")
        sys.exit()    

def write_sales(sales):
    with open(FILENAME, "w", newline="") as file:
        #TO DO
        writer=csv.writer(file)
        for row in sales:
            writer.writerow(row)
    
def view_monthly_sales(sales):
    print()
    month_view = []
    print("The amount of sales per month are:\n")
    for row in sales:
        month_view.append(str(row[0]) + " " + "$" + str(row[1]))
    for item in month_view:
        print(item) 
    print()
        
def view_yearly_summary(sales):
    print()
    total = 0
    #TO DO
    for row in sales:
        total=total+ int(row[1])   
    # get count
    #TO DO
    amount = len(sales)
    print("The total amount of sales are:")
    print("$" + str(total))
    print()
    # calculate average
    #TO DO
    avg= total/amount
    # format and display the result
    avg=round(avg,2)
    #TO DO
    print("The average sales for the year are")
    print("$" + str(avg))
    print()

def edit(sales):
    print()
    names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    name = input("Three-letter Month: ")
    print()
    name = name.title()
    index = names.index(name)
    if index != -1:
        #TO DO
        new = input("Amount of sales for "+ name +": ")
        print()
        for row in sales:
            if row[0] == name:
                row[1]=new
        month_view = []
        for row in sales:
            month_view.append(str(row[0]) + " " + "$" + str(row[1]))
        for item in month_view:
            print(item)
        print()
    else:
        print("Invalid three-letter month.")
        print()
    write_sales(sales)

def display_menu():
    print("COMMAND MENU")
    print()
    print("monthly - View monthly sales")
    print("yearly  - View yearly summary")
    print("edit    - Edit sales for a month")
    print("exit    - Exit program")
    print()    

def main():
    print("Monthly Sales program")
    print()
    
    sales = read_sales()
    display_menu()    
    while True:
        command = input("Command: ")
        if command == "monthly":
            view_monthly_sales(sales)
        elif command == "yearly":
            view_yearly_summary(sales)
        elif command == "edit":
            edit(sales)
        elif command == "exit":
            break
        else:
            print("Not a valid command. Please try again.\n")
    print()
    print("Bye!")

if __name__ == "__main__":
    main()
```
